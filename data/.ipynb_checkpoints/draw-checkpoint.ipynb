```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def drow_maker(img, pos, types, optw=100, opth=100):
    drow = cv2.drawMarker(img, 
                          pos, 
                          color=(0, 255, 0), 
                          markerType=cv2.MARKER_CROSS, 
                          markerSize=20, 
                          thickness=10, 
                          line_type=cv2.LINE_8)
    drow = cv2.putText(drow, types + ' as W:{} H:{}'.format(pos[0], pos[1]), (pos[0]+optw, pos[1]+opth), cv2.FONT_HERSHEY_SIMPLEX, 2.0, (0, 255, 255), thickness=5)
    plt.imshow(drow),plt.show()

    
left = cv2.imread("left.png")
right = cv2.imread("right.png")
print(left.shape, right.shape)
plt.imshow(left),plt.show()
plt.imshow(right),plt.show()
```

```python
drow_maker(right.copy(), (200, 630), types='R')
drow_maker(left.copy(), (180, 680), types='L')
```

```python
drow_maker(right.copy(), (360, 660), types='R')
drow_maker(left.copy(), (330, 710), types='L')
```
